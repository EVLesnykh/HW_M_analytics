**Исходные данные:**

Представим, что вы — разработчик пяти приложений, каждое из которых плюс-минус одинаково продвигается с точки зрения маркетинга. Предположим, у вас есть полные данные о показах и кликах (установках и всём, что после, разумеется тоже).<br>
Вам нужно ответить на 3 вопроса:<br>
1. Как детектировать Ad Stacking (подвид Click Spam)?<br>
2. Как детектировать Click Hijacking (клик вставляется после другого клика)?<br>
3. На какие виды фрода в первую очередь повлияет iOS 14.5+ и как именно?<br>

**Ответы:**
1. Ad Stacking. Существует альтернативная версия Click-спама. По метрикам конверсии она показывает чуть меньшие аномалии, но не менее успешно ворует органику. Как она работает?
Вы, как приложение, в котором крутится реклама, честно совершаете рекламные показы. Но в момент показа или клика вместо того, чтобы совершить переход по одной трекинговой ссылке, вы совершаете множество скрытых переходов по всем известным вам ссылкам (полученным от рекламной сети).
Теперь, если пользователь будет устанавливать не только ту игру, которой он совершенно честным образом заинтересовался, но и позже, органически какую-то ещё из рекламируемых в сети, а время между кликом и установкой будет меньше окна атрибуции, эта установка будет атрибутирована мошенническому источнику.<br>
**Такой фрод значительно проще детектировать не in-house, а на стороне трекера, ведь MMP в отличие от вас видит все клики, которые через него проходят**.<br>
2. Ещё один вид мошенничества для борьбы с которым используется индекс мультиканальности и мультиканальная атрибуция —  Click Hijacking. Механика схожа с Install Hijacking, но здесь вредоносное приложение обнаруживает реальный клик и посылает отчёт о фальшивом клике из конкурирующей сети, перехватывая таким образом клик и саму установку.
В модели от Appsflyer есть последний клик, который конвертит, и первый contributor — предыдущий клик в воронке. Соответственно, виден паттерн в мультиканальной атрибуции: предпоследний клик неестественно близок к последнему. Сразу можно отсекать такой скачок и работать с этими данными с подозрением на Click Hijacking.<br>
3. С момента выхода iOS 14.5 случаи мошенничества с использованием ботов и кликов участились. Дело в том, что в результате обновления iOS сократилось количество информации, передаваемой после взаимодействия пользователя с рекламой. Среди изначально расплывчатых данных клик-фермам чаще удаётся засчитать фальшивую атрибуцию.<br>
