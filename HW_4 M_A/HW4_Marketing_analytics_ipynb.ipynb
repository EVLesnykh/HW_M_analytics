```python
final_df= pd.merge(hw3, df1,on='Country Code')
final_df.head()
```

```python
final_df.shape
```

```python
final_df.isna().sum()
```

```python
final_df['Spend'] = final_df['Amount Spent (USD)']
final_df.head()
```

```python
final_df['Profit'] = final_df['Revenue'] - final_df['Spend']
final_df.head()
```

```python
final_df['CPI'] = round((final_df['Spend'] / (final_df['Mobile App Installs']+final_df['Installs'])), 2)
final_df.head()
```

```python
final_df['ROI'] = ((final_df['Revenue'] - final_df['Spend'])/ final_df['Spend']) * 100
final_df.head()
```
